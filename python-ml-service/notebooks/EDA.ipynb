```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Karena file ini ada di folder notebooks/, kita naik satu level pake '../' ke folder data/
df = pd.read_csv('../data/trash_data.csv')

print("Jumlah data:", len(df), "baris")
print("Tampilan 5 Data Pertama:")
display(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['volume_sekarang'], bins=30, kde=True, color='teal')
plt.title('Distribusi Volume Sampah di Seluruh Kota')
plt.xlabel('Persentase Kepenuhan Tong Sampah (%)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='pickup_priority', order=['Low', 'Medium', 'Urgent', 'Critical'], palette='YlOrRd')
plt.title('Proporsi Status Urgensi Penjemputan Truk Sampah')
plt.xlabel('Tingkat Prioritas')
plt.ylabel('Jumlah Tong Sampah')
plt.show()
```
